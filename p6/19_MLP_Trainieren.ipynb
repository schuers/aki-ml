# MLP Modelle trainieren

In diesem Aufgabenblatt geht es darum, [MLP](https://de.wikipedia.org/wiki/Multi-Layer-Perzeptron)-Modelle mit *Scikit-learn* und *Keras* aufzustellen und zu trainieren.
Wir verwenden dazu einen synthetischen, nicht-linear-separierbaren Datensatz mit 3 Klassen von Datenpunkten.

```python
import numpy as np
import matplotlib.pyplot as plt

def make_spirals(N=100, dim=2, classes=2, random_state=0):
    """ Generate a spiral dataset.
    Parameters
    ----------
    N : int
        Number of points per class.
    dim : int
        Number of dimensions.
    classes : int
        Number of classes.
    random_state : int
        Random seed for reproducibility.

    Returns
    -------
    X : ndarray, shape (N*classes, dim)
        The generated data.
    y : ndarray, shape (N*classes,)
        The labels for the data.
    """

    np.random.seed(random_state)
    X = np.zeros((N*classes,dim))            # train data
    num_train_examples = X.shape[0]
    y = np.zeros(N*classes, dtype='uint8')   # labels
    for j in range(classes):
        ix = range(N*j,N*(j+1))
        r = np.linspace(0.0,1,N)             # radius
        k = classes+1
        t = np.linspace(j*k,(j+1)*k,N) + np.random.randn(N)*0.2 # theta
        X[ix] = np.c_[r*np.sin(t), r*np.cos(t)]
        y[ix] = j
    return X, y

N=60            # number of points per class
dim=2           # number of dimensions
classes=3       # number of classes
random_state=0  # random seed for reproducibility

# Generate the dataset
X, y = make_spirals(N, dim, classes, random_state)

# Plot the dataset
fig = plt.figure()
plt.title("Spiral Dataset")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.grid()
plt.tight_layout()

# Scatter plot
plt.scatter(X[:, 0], X[:, 1], c=y, s=40, cmap=plt.cm.Spectral)
plt.xlim([-1,1])
plt.ylim([-1,1])
```

Wie man in folgendem Beispiel sieht, schneidet ein Lineares Modell bei diesen Daten sehr schlecht ab:

```python
from sklearn.linear_model import LogisticRegression
from sklearn.multiclass import OneVsRestClassifier

# Erstelle das Modell mit OneVsRestClassifier
clf = OneVsRestClassifier(LogisticRegression(random_state=0)).fit(X, y)
score = clf.score(X, y)

print("Accuracy = %.2f%%" % (score*100))
```

```python
def plot_decregions_2d(X,clf):
    '''
    Plotte Entscheidungsgrenzen in einem
    Datensatz mit zwei Attributen
    
    Parameter:
    X: 2d numpy array der Groesse (m,n)
    clf: Klassifizierer mit einer Methoder "predict"
    '''
    cmap = plt.get_cmap('Set1', 3)
    h = 0.02
    x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
    y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1
    xx, yy = np.meshgrid(np.arange(x_min, x_max, h),
                         np.arange(y_min, y_max, h))
    Z = clf.predict(np.c_[xx.ravel(), yy.ravel()])
    print(Z.shape)
    if Z.ndim>1:
        Z = np.argmax(Z, axis=1)
        print(Z.shape)
    Z = Z.reshape(xx.shape)
    fig = plt.figure()
    plt.contourf(xx, yy, Z, cmap=cmap, alpha=0.4)
    plt.scatter(X[:, 0], X[:, 1], c=y, s=40, ec='black', cmap=cmap)
    plt.xlim(xx.min(), xx.max())
    plt.ylim(yy.min(), yy.max())
    plt.title("2-D decision regions for dataset")
    plt.xlabel("Feature 1")
    plt.ylabel("Feature 2")

plot_decregions_2d(X,clf)
```

**Aufgabe: Erstellen Sie das gleiche Modell mit [Tensorflow](https://www.tensorflow.org/) bzw [Keras](https://keras.io/).**

Verwenden Sie dazu die [*Keras Sequential API*](https://keras.io/api/models/sequential/).

```python
import tensorflow as tf

# Konvertiere X und y zu float64
tX = X.astype('float64')
ty = y.astype('float64')

optimizer ='RMSprop'
loss = 'sparse_categorical_crossentropy' # Wieso? Ist die selbe die im Buch in Kapitel S. 357 für eine Klassifikation verwendet wurde
metrics = ['accuracy'] # Wieso? Ist die selbe die im Buch in Kapitel S. 357 für eine Klassifikation verwendet wurde

model = tf.keras.Sequential()

model.add(tf.keras.layers.Input(shape=[2]))
model.add(tf.keras.layers.Dense(100, activation='relu')) # MLPClassificator has one hidden layer with 100 neurons and uses relu as the default acitvation layer. 
model.add(tf.keras.layers.Dense(classes, activation='softmax')) # Add an output neuron for each class. Use the softmax function. 
# Modell erzeugen
model.compile(optimizer=optimizer, loss=loss, metrics=metrics)

model.summary()
```

```python
#Modell trainieren
model.fit(tX, ty, epochs=500, verbose=0)
plot_decregions_2d(tX,model)
```

</div>

```python
eval = model.evaluate(tX, ty)

test_loss = eval[0]
test_acc = eval[1]


print('Test accuracy: %.2f%% ' % (test_acc * 100))
```
